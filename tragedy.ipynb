# **Story**

## Some say there's plenty of fish in the sea, but I beg to differ. Here is yet another story, in which humans create tragedy through overconsumption.
## Let's introduce our first character: the Atlantic bluefin tuna.

```python
# create a new dataframe to remove any "irrelevant" countries - countries that distract from the story
mod_bluefin = df[(df.Species == 'Atlantic bluefin tuna')&(df.Country != 'Ireland')]

# create a line graph for atlantic bluefin tuna
ax=sns.lineplot(x = "Year", y = "Catch (Tons)", data = mod_bluefin, hue='Country', ci = False)
plt.title("Catch of Atlantic Bluefin Tuna Over Time", size=20)
plt.xlabel(None)
plt.ylabel("Catch (tons)")
ax;
```

### In the 1950's, the Atlantic bluefin tuna was a very popular catch in Norway. Each time it was overfished (in 1952, 1954, and 1962), the population of tuna understandably dropped. 
### Canada and the US did not catch as much tuna as Norway. However, similar to Norway, there has been spikes in the catch of tuna (in the early 1960s and 1970) followed by a fall.
### Since the craze of tuna 50-70 years ago, the Atlantic bluefin tuna is caught much less often, if at all, which is likely due to established regulations and international cooperation.

## Next, we have the Atlantic cod. 

```python
# create a new dataframe to remove any "irrelevant" countries
mod_cod = df[(df.Species == 'Atlantic cod')&(df.Country != 'Ireland')&(df.Country != 'Portugal')&(df.Country != 'Ireland')&(df.Country != 'Greenland')]

# create a line graph for atlantic cod
ax=sns.lineplot(x = "Year", y = "Catch (Tons)", data = mod_cod, hue='Country', ci = False)
plt.title("Catch of Atlantic Cod Over Time", size=20)
plt.xlabel(None)
plt.ylabel("Catch (tons)")
ax;
```

### Evidently, the Atlantic cod is a popular fish on both sides of the pond. The catch of cod is very tumultuously up and down in a number of countries, including Canada, Iceland, and Norway. 
### Currently, the population of the Atlantic cod is significantly below target population levels and so the fishing rate is reduced to end overfishing. Did Iceland (in orange) and Norway (in green) get the memo?
### The same trend of human overfishing repeats for the other fish species in this data set, which includes Atlantic halibut, Atlantic salmon, swordfish, and yellowfin tuna.

## Atlantic salmon:

```python
# create line graph for atlantic salmon
ax=sns.lineplot(x = "Year", y = "Catch (Tons)", data = salmon, hue='Country', ci = False)
plt.title("Catch of Atlantic Salmon Over Time", size=20)
plt.xlabel(None)
plt.ylabel("Catch (tons)")
ax;
```

### The history of Atlantic salmon is particularly tragic. As you can see, history of Atlantic salmon catches is particularly turbulent from 1950 to 1987. The outcome is unfortunately a complete collapse in Atlantic salmon. 
### As a result, commercial and recreational fishing for salmon is prohibited. Meanwhile, the salmon we find in supermarkets are farm-raised.

## Swordfish:

```python
# create a line graph for swordfish
ax=sns.lineplot(x = "Year", y = "Catch (Tons)", data = swordfish, hue='Country',ci = False)
plt.title("Catch of Swordfish Over Time", size=20)
plt.xlabel(None)
plt.ylabel("Catch (tons)")
ax;
```

### The Canadian demand for swordfish was greater than the American demand. There are sudden drops in the catch of swordfish as a result of overfishing in 1963, 1970, and 1979 in Canada. The drops in American catch are not as drastic. 
### Nevertheless, the catch of swordfish by Canada and the U.S. are similar today, which is presumably the result of effective intervention.

## Atlantic halibut:

```python
# create a new dataframe to remove any "irrelevant" countries
mod_halibut=df[(df.Species == 'Atlantic halibut')&(df.Country != 'United States of America')]

# create a line graph for halibut
ax=sns.lineplot(x = "Year", y = "Catch (Tons)", data = mod_halibut, hue='Country', ci = False)
plt.title("Catch of Atlantic Halibut Over Time", size=20)
plt.xlabel(None)
plt.ylabel("Catch (tons)")
plt.legend(loc='upper right')
ax;
```

### Apart from Canada and Norway, there is a general downward trend of Atlantic halibut catches over time, which reflects the halibut’s loss of population. 
### Similar to the Atlantic cod, there is a lack of international cooperation to allow for the repopulation of Atlantic halibut. The culprits this time are Canada and Iceland. (I’m beginning to see a behavioral pattern regarding Iceland.)

## Our last character is Yellowfin tuna.

```python
# create a line graph for yellowfin tuna
ax=sns.lineplot(x = "Year", y = "Catch (Tons)", data = yellowfin, hue='Country', ci = False)
plt.title("Catch of Yellowfin Tuna Over Time", size=20)
plt.xlabel(None)
plt.ylabel("Catch (tons)")
ax;
```

### Yellowfin tuna was popular in the mid-1980’s and fishermen avidly caught them to feed the American appetite. There were two spikes of overfishing followed by immediate depletion in 1988 and 1992, but since then, the catch of yellowfin tuna has decreased significantly. 
### This tuna is now sustainably managed and responsibly harvested. With such measures in place, yellowfin tuna can avoid the tragedy of the commons.

## It’s hard to be optimistic about human greed, but I can see some hope. There are organizations, like NOAA, who oversee the sustainable management and harvest of fish. There is also fishermen of many nationalities, who abide by the established regulations. Apart from some outliers (Iceland, I'm looking at you), most countries are cooperating. 
## The populations of swordfish and yellowfish tuna are the most likely to be safe. Plans are in place to rebuild populations of more endangered fish species, like the Atlantic salmon and Atlantic halibut.
